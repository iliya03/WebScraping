```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
#Loading the webpage content
r= requests.get("https://www.iplt20.com/auction")

#soup is the object of Beautiuful Soup
soup = bs(r.content)

print(soup.prettify())
```

```python
Title = soup.find_all("title", attrs={"id": "page_title"})
Title
```

```python
body = soup.find('body')
div= body.find('div')
div
```

```python
#searching the specific string
#name = soup.find_all('title', string='Indian Premier League Official Website')
import re
name = soup.find_all('title', string = re.compile('Indian'))
name
```

Lets Analyse Ipl 2024 Auction Data of Ipl Teams





```python
#table_element = soup.find('table', class_='ih-pcard-sec')
table = soup.find('table', class_ = 'ih-td-tab auction-tbl', id='t1')
table
```

```python
table_header= table.find_all('th')
table_header
```

```python
#extracting the header text
header = []
for a in table_header:
  name= a.text
  header.append(name)

df= pd.DataFrame(columns=header)
print(df)
```

```python
rows = table.find_all('tr')
rows
```

```python
table_header = table.find_all('th')

header_names = [c.string for c in table_header]

table_data = table.find('tbody').find_all('td')

l=[]
for tr in rows:
  td = tr.find_all('td')
  table_rows = [tr.get_text() for tr in td]
  l.append(table_rows)

df = pd.DataFrame(l, columns=header_names)
df=df.drop(df.index[0])
df
```
