# Web Scraping real website using pandas

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
r= requests.get("https://en.wikipedia.org/wiki/List_of_the_largest_software_companies")

#soup is the object of Beautiuful Soup
soup = bs(r.content)

print(soup.prettify())
```

```python
table = soup.find('table', class_ = 'wikitable sortable plainrowheads')
print(table)
```

```python
soup.find_all('table')[6]
#soup.prettify()
```

```python
table_headers = table.find_all('th')
table_headers
```

```python
#extracting the table headers as text
table_title = [title.text.strip() for title in table_headers]
print(table_title)
```

```python
#putting the titles into a dataframe
df = pd.DataFrame(columns=table_title)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    each_data= [each_data.text.strip() for each_data in row_data]
    print(each_data)
    
    length = len(df)
    df.loc[length] = each_data
```

```python
df
```

```python
#another method using append and empty list object 'data'

data = []
for row in column_data[1:]:
    row_data = row.find_all('td')
    each_data = [each.text.strip() for each in row_data]
    data.append(each_data)

df = pd.DataFrame(data=data, columns=table_title)
df
```
